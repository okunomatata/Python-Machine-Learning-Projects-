## Prepare the Data for ML (3 points) 

**(5) Split the data into train and test sets (random sate = 42, test_size = .20)**

*Import train_test_split*

```python
# from sklearn.
from sklearn.model_selection import train_test_split
```

```python
# X_train, X_test, y_train, y_test = 
X_train, X_test, y_train, y_test = train_test_split(x, y, test_size=.40, random_state = 42)
```

```python
X_train
```

**(6) Scale the training and testing data (following the Lecture 1 notebook example)**

```python
# X_train_scaled
# X_test_scaled
from sklearn.preprocessing import StandardScaler
X_train = X_train.iloc[:, 0:10] #not sure why we did this
s = StandardScaler()
X_train_scaled = pd.DataFrame(s.fit_transform(X_train), columns = X_train.columns)
X_train_scaled
```

```python
X_test = X_test.iloc[:, 0:10]
s = StandardScaler()
X_test_scaled = pd.DataFrame(s.fit_transform(X_test), columns = X_test.columns)
X_test_scaled
```

## Tune and Evaluate Model (4 points)

**(7) Import the KNeighborsRegressor**

```python
# from sklearn. ...
from sklearn.neighbors import KNeighborsRegressor
```

**(8) Modify the code block below to compute two performance metrics (e.g. MAE and R2)**

```python
from sklearn.metrics import mean_squared_error, r2_score
results = {}
for i in np.linspace(3, 25, 5, dtype = int):
    knn_model = KNeighborsRegressor(n_neighbors=i)
    knn_model.fit(X_train_scaled, y_train)
    pred_dance = knn_model.predict(X_test_scaled)
    results[i] = knn_model.score(X_test_scaled, y_test), mean_squared_error(y_test, pred_dance) 
pd.DataFrame.from_dict(results, orient = "index", columns = ["R2", "MAE"])

```

```python
#Trying a different method to get MAE and R2
pred_dance = knn_model.predict(X_test_scaled)
from sklearn.metrics import mean_squared_error, r2_score
results = {'MAE': mean_squared_error(y_test, pred_dance),
           'R2': r2_score(y_test,pred_dance)
          }
results
```

**(9) Describe the model performance in the cell below**

As we increase the number of neightbers used in the algorithm we can see model improvement. Our Pearson R Squared Coefficient goes up and our Mean Average Error goes down. 

**Improvement**

Although we have not formally introduced the k-NN algorithm (see **"Notes"** for more clarification), Lecture 1 discussed bias and variance and their role in machine learning, particularly in optimizing models for our data/problem. For instance, we want models/algorithms that are not overly simple as we will not be able to capture the trends in the data (often called the signal). Yet we want to avoid excessive model complexity as we will begin capturing too much noise. Below, look at the documentation for KNeighborsRegressor and the weights parameter. Paste your code from above, changing the parameter value from the default, "uniform," to "distance."

**Notes**: k-NN is in a family of algorithms that rely on distance metrics to generate predictions. Data points are projected into a feature space and observations (here songs) are either near or far relative to each other and this naturally leads to the concept of "Neighbors." A prediction is possible when projecting a new data point into this space. Next, take the average danceability score from "k" neighbors. This is the prediction. Therefore, the n_neighbors parameter is setting this cutoff for the number of neighbors (songs) that are going into the prediction. The weights parameter slightly alters how neighbors contribute to the prediction.

**(10a) Paste the k-NN code below, setting the weights parameter to distance**

```python
results = {}
for i in np.linspace(3, 25, 5, dtype = int):
    knn_model = KNeighborsRegressor(n_neighbors=i, weights = "distance")
    knn_model.fit(X_train_scaled, y_train)
    results[i] = knn_model.score(X_test_scaled, y_test)
pd.DataFrame.from_dict(results, orient = "index", columns = ["R2"])
```

**(10b) Interpret the model performance (refer to the documentation as to what this parameter does)** 
To guide you, this distinction in performance is related to global vs. local solutions (in how the algorithm is 
predicting, in this case, "danceability"), and ties into our discussion on variance, bias, and model complexity. Use some of these terms in comparing this result to the previous one.

We were underfitting the model before. By changing our weights parameter we add some complexity to
our model, which means we are reducing bias and increasing variance.
